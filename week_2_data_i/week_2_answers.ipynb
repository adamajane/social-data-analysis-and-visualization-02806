# Part 2: Merging Data

```python
import pandas as pd

crime_data_new = pd.read_csv(
    "../data/Police_Department_Incident_Reports__2018_to_Present_20260203.csv"
)

crime_data_old = pd.read_csv(
    "../data/Police_Department_Incident_Reports__Historical_2003_to_May_2018_20260210.csv"
)

print("Datasets loaded")
```

## Mini Exercises (Steps 2.1 – 2.3)

```python
# Schema exploration
columns_new = crime_data_new.columns
columns_old = crime_data_old.columns

print("New crime dataset columns:\n", columns_new.tolist(), "\n")
print("Old crime dataset columns: \n", columns_old.tolist(), "\n")

print("Identical columns: \n", columns_new.intersection(columns_old), "\n")

# Columns only in crime new
only_crime_new = columns_new.difference(columns_old)
print("Columns unique to new crime data: \n", only_crime_new, "\n")

# Columns only in crime old
only_crime_old = columns_old.difference(columns_new)
print("Columns unique to old crime data: \n", only_crime_old, "\n")
```
